### Se pide:

A partir del archivo "USA_Housing.csv", realizar la regresión lineal entre cada columna numérica con la columna 'price'. 

Cuando existe una columna(variable) que es la que se quiere comparar, se le llama 'target'. 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd

df = pd.read_csv("../data/USA_Housing.csv")
df
```

```python
X = df[['Avg. Area Income']]
```

```python
y = df["Price"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
```

```python
lin_reg.score(X_test, y_test) * 100
```

```python
lin_reg.score(X_train, y_train) * 100
```

```python
plt.scatter(X, y, color='c')
plt.plot(X, lin_reg.predict(X), color='b')

plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
```

```python
def linear_reg(column): 
    X = df[[column]]
    y = df["Price"]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
    lin_reg.fit(X_train, y_train)
    print("Score test:",lin_reg.score(X_test, y_test) * 100)
    print("Score train:", lin_reg.score(X_train, y_train) * 100)

    plt.scatter(X, y, color='c')
    plt.plot(X, lin_reg.predict(X), color='b')
    return plt.show()
```

```python
linear_reg('Avg. Area Income')
```
