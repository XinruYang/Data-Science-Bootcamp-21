Say the number generated by the computer is 1038. An example interaction could look like this:

  Welcome to the Cows and Bulls Game! 
  
  >Enter a number:

  >> 1234

  >>> 2 cows, 0 bulls

  >> 1253

  >>> 1 cow, 1 bull
  
  ...
Until the user guesses the number or write "STOP"

```python
# B
def cows_bulls():
    mi_numero = []

    while len(mi_numero) < 4:
        numero = str(random.randint(0,9))
        if numero not in mi_numero:
            mi_numero.append(numero)
    print(mi_numero)
    cow = 0
    contador = 0
    while cow < 4:
        numero_adivinar = input('Ingrese un número de 4 cifras sin repetir dígitos. Escribe STOP si quieres salir:')
        numero_adivinar_set = set(numero_adivinar)
        if (len(numero_adivinar_set) != 4 or not numero_adivinar.isdigit()) and numero_adivinar != 'STOP':
            print('Ingrese un número de 4 cifras distintas')
            print('======================')
            continue
        if numero_adivinar == 'STOP':
            print('Has abandonado el juego')
            return None
        numero_adivinar_lista = list(numero_adivinar)
        cow = 0
        bull = 0
        for i,elem in enumerate(mi_numero):
            if elem == numero_adivinar_lista[i]:
                cow += 1
            elif elem in numero_adivinar_lista:
                bull += 1
        contador += 1
        print('Intento número:', contador)
        print('-----------------')
        print('Número elegido:', numero_adivinar)
        print('cows:', cow)
        print('bulls:', bull)
        print('======================')
    print('GANASTE !!')
    print('Lo has conseguido en ', contador, ' intentos!')

cows_bulls()
```

```python
# D 

import random 

# Un número es generado de manera aleatoria
generated_num = str(random.randint(1000,9999))

acum_cows = 0
guess_str =""
while acum_cows < 4 and guess_str != "STOP":
    # los acumuladores se inicializan con cada iteración
    acum_cows = 0
    acum_bulls = 0

    guess_str = input("Please enter a number between 1000 and 9999 (STOP to stop): ")
    print(f"Your number {guess_str}")
    # variable auxiliar para poder modificar los dígitos si se encuentran vacas
    generated_num_cowed = list(generated_num) 

    # En el primer recorrido del bucle, se obtienen los números que coinciden y se sustituyen los dígitos por una 'c' para no ser contabilizados en la búsqueda de toros
    for i, item in enumerate (generated_num):
        if item == guess_str[i]:
            acum_cows += 1
            generated_num_cowed[i] = "c"

    # En el segundo recorrido, se obtien los números que están pero no coinciden
    for i in range(len(generated_num_cowed)):
        for j in range(len(guess_str)):
                if guess_str[i] == generated_num_cowed[j]:
                    acum_bulls += 1


    print(f"{acum_cows} cows, {acum_bulls} bulls.")
```

```python
# C
# -------- Modules --------
import random
from functools import reduce

# -------- Function to check number of cows/bulls in guess --------
def checker(num, guess):
    num = str(num)
    guess = str(guess)

    cows = 0    # Value and position
    bulls = 0   # Position

    for i, x in enumerate(num):
        if x in guess:
            bulls += 1

            if guess[i] == x:
                bulls -= 1
                cows += 1

    return [bulls, cows]

# -------- To create the random number to guess --------
def random_non_repetitive():
    x = []
    count = 0

    while count < 4:
        nr = random.randint(0, 9)
        if nr not in x:
            x.append(str(nr))
        count += 1

    num = int(reduce(lambda x, y: x + y, x))
    return num

# -------- To play the game --------
# I have to pass it a num. In case I want to have a random number, then I pass the function above
def game(num):
    bulls, cows = [0, 0]

    while cows < 4:
        guess = input('Enter your guess: ')
        if guess == 'STOP':
            break
        print(guess)
        bulls, cows = checker(num, guess)
        print('{} cows, {} bulls'.format(cows, bulls))

    print('Your guess: ', guess)
    print('Your num: ', num)

# numero a adivinar
num = random_non_repetitive()
print("A adivinar:", num)
game(num)
```
