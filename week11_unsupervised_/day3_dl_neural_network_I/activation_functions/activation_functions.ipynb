### ReLu

```python
def ReLU(x):
    return (abs(x) + x) / 2
```

```python
ReLU(-35252)
```

### Sigmoid

```python
import math
# Valor entre 0 y 1
def sigmoid(x):
  return 1 / (1 + math.exp(-x))
```

```python
sigmoid(10)
```

### Hyperbolic tangent

```python
import tensorflow as tf 
# Entre -1 y 1
tf.keras.activations.tanh(-12412.0)
```
